### Import necessary packages
Here the package matplotlib is imported in a different way sdue to a bug

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
from natsort import natsorted, ns
import os
import holoviews as hv
hv.notebook_extension('matplotlib')
```

```python
dirList = os.listdir("dataFiles2/")
dirList = natsorted(dirList)
File_Path = ["dataFiles2/" + s for s in dirList]
```

```python
Data= [np.loadtxt(f) for f in File_Path]
```

```python
x=np.linspace(0,15,np.shape(Data)[1])
fig = plt.figure()
ax = plt.axes(xlim=(0, 15), ylim=(0, 2))
line, = ax.plot([], [], lw=2)
```

```python
def init():
    line.set_data([], [])
    return line,

def ThinFilmAnime(i):
    global Data, x
    line.set_data(x, np.array(Data[i]))
    return line,
```

```python
anim = matplotlib.animation.FuncAnimation(fig,ThinFilmAnime, init_func=init,
                               frames=np.shape(Data)[0], interval=1, blit=True)
np.shape(anim)
anim.save('filmFestival2.mp4', fps=300, extra_args=['-vcodec', 'libx264'])
```
